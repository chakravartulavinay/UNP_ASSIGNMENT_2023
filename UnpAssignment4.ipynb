```python
 # 7. Find the city with the highest average humidity and display the city name and the average humidity value.
avg_humidity=df.groupby('City').agg({'Humidity (%)':'mean'}).sort_values('Humidity (%)')
avg_humidity_dataframe=pd.DataFrame(avg_humidity)
max_avg_humidity=avg_humidity_dataframe['Humidity (%)'].max()
result = avg_humidity_dataframe.loc[avg_humidity_dataframe['Humidity (%)']==max_avg_humidity]
result
```

```python
# 8. Create a new DataFrame that includes only the rows where the 'Condition' column is 'Rainy' and the 'Wind Speed (km/h)' is greater than 10.
rainy_days=df.loc[df['Condition']=='Rainy']
wind_speed=df.loc[df['Wind Speed (km/h)'] > 10]
d=pd.merge(rainy_days,wind_speed)
d
```

```python
# 9. Calculate the percentage of rainy days with wind speed greater than 10 km/h.
d['percent']=(d['Wind Speed (km/h)']/d['Wind Speed (km/h)'].sum())*100
d
d.loc[:,['Condition','Wind Speed (km/h)','percent']]
```
